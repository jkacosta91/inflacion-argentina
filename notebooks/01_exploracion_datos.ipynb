```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos
df = pd.read_csv('../data/datos_macro.csv', parse_dates=['periodo'])
```

```python
# Mostrar primeras filas
print("üìå Primeras filas del dataset:")
display(df.head())
```

```python
# Info general
print("\nüìã Informaci√≥n general del dataset:")
df.info()
```

```python
# Valores faltantes
print("\n‚ùì Valores faltantes por columna:")
print(df.isnull().sum())
```

```python
# Estad√≠sticas descriptivas
print("\nüìä Estad√≠sticas descriptivas:")
display(df.describe())
```

```python
# Correlaci√≥n entre variables
print("\nüìà Matriz de correlaci√≥n:")
corr = df.drop(columns='periodo').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de Correlaci√≥n entre Variables Econ√≥micas")
plt.show()
```

```python
# Visualizar series temporales
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12), sharex=True)
fig.suptitle("üìà Series Temporales Econ√≥micas (2017-2024)", fontsize=18, y=0.94)

columnas = [
    'IPC_General', 'EMAE', 'Tasa_Desempleo', 'RIPTE',
    'Tipo_de_Cambio_USD', 'Tipo_de_Cambio_Multilateral'  # sin 'Base_Monetaria_Total'
]
for i, col in enumerate(columnas):
    ax = axes[i // 2, i % 2]
    ax.plot(df['periodo'], df[col], label=col)
    ax.set_title(col)
    ax.set_ylabel("Valor")
    ax.grid(True)
    ax.legend()

for ax in axes[-1]:
    ax.set_xlabel("Fecha")

plt.tight_layout(rect=[0, 0, 1, 0.95])  # deja espacio para el t√≠tulo superior
plt.show()
```
